```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
item_list = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
te = TransactionEncoder()
te_ary = te.fit(item_list).transform(item_list)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
fpgrowth(df,min_support=0.6)
```

```python
freq_itemsets = fpgrowth(df, min_support=0.6, use_colnames=True)
freq_itemsets['length'] = freq_itemsets["itemsets"].apply(lambda x : len(x))
freq_itemsets
```

```python
rules = association_rules(freq_itemsets, min_threshold=0.8)
rules
```
