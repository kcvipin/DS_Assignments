```python
import pandas as pd
import numpy as np

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
book.rename({'User.ID':'User_Id','Book.Title':'Title','Book.Rating':'Rating'},axis=1,inplace=True)
```

```python
item_book_df = book.pivot_table(index='Title',
                                 columns='User_Id',
                                 values='Rating')
```

```python
item_book_df
```

```python
item_book_df.fillna(0,inplace=True)
item_book_df
```

```python
item_similarity = 1 - pairwise_distances( item_book_df.values,metric='cosine')
```

```python
item_similarity
```

```python
item_sim_df = pd.DataFrame(item_similarity)
item_sim_df.columns = book['Title'].unique()
item_sim_df.index = book['Title'].unique()
item_sim_df
```

```python
item_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(item_similarity, 0)
item_sim_df.iloc[0:5, 0:5]
```
